## FILTROS


   ## Filtro temática y valoración


```python
def filtro_valoracion_tematica(df, sesiones):
    tematica_unique_list = df['Temática'].unique()

    filtro_tematica_df = pd.DataFrame()
    for tematica in tematica_unique_list:
        df_tematica= df[(df['Nº Sesiones'] <= int(sesiones)) & (df['Temática'] == tematica)]
        filtro_tematica_df = pd.concat([filtro_tematica_df, df_tematica.head(2)])  # Obtener dos filas por temática
        print("filtro_tematica_df ---> ", tematica)
    
    recomendaciones_valoracion_tema_df = filtro_tematica_df.drop_duplicates().sort_values(by='Valoración', ascending=False)
    
    return recomendaciones_valoracion_tema_df
```

```python
recomendacion_2 = filtro_valoracion_tematica(df_merge, 5)
recomendacion_2
```
